```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("/Users/yashpatel/Personal Projects /Hearing Wellness/Main/HearingData.xlsx")
print(df)
```

```python
plt. figure(figsize=(10,8))
ax = (sns.countplot(x="Daily_Headphone_Use", hue='Daily_Headphone_Use',data=df))
ax.set(ylabel="Number Of People")
ax.set(xlabel="Daily Headphone Usage")
ax.set_title("Daily-Headphone-Usage-Survey")
plt.savefig("Daily-Headphone-Usage-Survey.png")
```

```python
plt. figure(figsize=(10,8))
ax = (sns.countplot(x="Hearing_FOMO", hue='Hearing_FOMO',data=df))
ax.set(ylabel="Number Of People")
ax.set(xlabel="Hearing_FOMO")
ax.set_title("Hearing-FOMO-Data")
plt.savefig("dailyHeadphoneUsageSurvey.png")
```
